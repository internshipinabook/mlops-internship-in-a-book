## 🚀 CAPSTONE EXTENSION — "Ship It" (Optional, Public Portfolio Track)

*Not required for the core 12-week path. The core capstone (Monday-Friday above) stays
internal to DataFlow Infrastructure, exactly as scoped. This extension is for anyone who wants
a public, linkable portfolio artifact at the end of the internship.*

### Why this is a separate track, not a core requirement

Deploying a model publicly changes what data is safe to use. `CreditDefault-NG` (used
throughout Weeks 0–11) is the real, anonymised UCI Credit Card Default dataset wearing a
Nigerian-fintech narrative wrapper — appropriate for local coursework, not for a live public
endpoint. This extension trains and serves on a **fully synthetic look-alike dataset** instead
— same 23-feature schema, same value ranges, zero real records.

### What you'll build

1. **Generate the synthetic dataset** (already schema-matched to everything you built in
   Weeks 1–11 — same column names, same PAY_0 bureau codes, same LIMIT_BAL/BILL_AMT/PAY_AMT
   structure):
   ```bash
   python generate_synthetic_credit_data.py
   # produces creditdefault_synthetic_{training,holdout,current}.csv
   ```
2. **Train and register** a model on the synthetic training set using your Week 1 MLflow
   pipeline, unchanged.
3. **Deploy publicly** using your Week 8 FastAPI + Docker server, to a free tier of your
   choice (Render, Hugging Face Spaces, Railway, Fly.io).
4. **Write the model card and README** — this is the part that matters most for a public
   artifact. It must state, prominently, near the top of the README:

   > **This model is trained entirely on synthetic data structurally modelled on the UCI
   > Credit Card Default dataset. No real cardholder records are used, stored, or exposed by
   > this demo. Predictions from this endpoint must not be used for any real credit decision.**

5. **Submit:**
   - Public GitHub repository (code + synthetic data generator + README + model card)
   - Live endpoint URL (`/health` and `/predict` reachable)
   - One-page portfolio summary: what you built, which weeks' skills it demonstrates, and a
     link to a short demo (a `curl` transcript is enough — a screen recording is a bonus, not
     a requirement)

### Common Mistakes (Ship It track)

**Deploying the real UCI-derived data publicly "because it's just anonymised numbers anyway."**
Anonymised is not the same as synthetic — anonymised data still describes real individuals'
financial histories. Public-facing deployment is exactly the context where that distinction
matters most.

**Skipping the model card because "it's just a demo."** A public demo without a model card
is indistinguishable, to a stranger finding it online, from a real credit-scoring tool. The
disclaimer is not optional politeness — it's the thing that keeps the demo honest.

**Forgetting to pin the generator's seed.** If `generate_synthetic_credit_data.py` isn't
run with a fixed seed, two people cloning your repo get two different datasets, and your
"reproducible" portfolio piece silently isn't.


## Common Mistakes

**Building before writing the ADRs:** A pipeline without ADRs cannot be maintained by anyone not present for its construction.

**PRR as a box-checking exercise:** A PRR where every item is PASS because the engineer filled it in without checking is a documentation failure.

**A letter that only describes what works:** The most useful letter describes what will break first. Every pipeline has a known failure mode. Document it explicitly.
